```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('MBA.csv')
```

```python
df.info()
```

```python
# load first few rowa
df.head()
```

```python
df.isnull().sum()
```

```python
df['race'] = df['race'].fillna('Unknown')
df['admission'] = df['admission'].fillna('Not Confirmed')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.drop('application_id',axis=1,inplace=True)
```

```python
#race count
plt.figure(figsize=(6,3))
sns.countplot(x='race',data=df,palette='Set1')
plt.show()
```

```python
#work_industry count
plt.figure(figsize=(6,3))
sns.countplot(y='work_industry',data=df,palette='Set1')
plt.show()
```

```python
#major count
plt.figure(figsize=(4,2))
sns.countplot(x='major',data=df,palette='Set1')
plt.show()
```

```python
#major count
plt.figure(figsize=(3,2))
sns.countplot(x='international',data=df,palette='pastel',)
plt.show()
```

```python
plt.figure(figsize=(5,2))
sns.countplot(x='admission',data=df,palette='deep')
plt.show()
```

```python
plt.figure(figsize=(3,2))
sns.countplot(x='gender',data=df,palette='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(4,2))

crosstab = pd.crosstab(df['gender'],df['international'])
crosstab.plot(kind='bar')
plt.xticks(rotation=0)
plt.show()
```

```python
crosstab = pd.crosstab(df['gender'],df['major'])
crosstab.plot(kind='bar')
plt.xticks(rotation=0)
plt.show()
```

```python
crosstab = pd.crosstab(df['gender'],df['race'])
crosstab.plot(kind='bar')
plt.xticks(rotation=0)
plt.show()
```

```python
crosstab = pd.crosstab(df['gender'],df['work_industry'])
crosstab.plot(kind='bar',figsize=(12,6))
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(7,3))
sns.boxplot(x='work_industry',y='gmat',data=df)
plt.xticks(rotation=90)
plt.show()
```
